```python
def convert_to_months(value):
    if pd.isna(value):
        return np.nan
    parts = value.split(' ')
    years = int(parts[0])
    months = int(parts[3])
    total_months = years * 12 + months
    return total_months
```

```python
df['Credit_History_Age_Months'] = df['Credit_History_Age'].apply(convert_to_months)
```

```python
df['Credit_History_Age_Months']
```

```python
df['Credit_History_Age_Months'] = df['Credit_History_Age_Months'].fillna(df['Credit_History_Age_Months'].mean())
```

```python
df.Amount_invested_monthly = df.Amount_invested_monthly.fillna(df.Amount_invested_monthly.mean())
```
